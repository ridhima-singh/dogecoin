```python
#Description: Predict the future price of a DOGECOIN using machine learning, facebook prophet and python
```

```python
#import pandas library
import pandas as pd
```

```python
#load the data
from google.colab import files
files.upload()
```

```python
#read in the data
df = pd.read_csv('DOGE-USD (1).csv')
#show the data
df
```

```python
#Select the date and the price
df = df[['Date', 'Close']]
#rename the features: 
df = df.rename(columns={'Date' : 'ds', 'Close' : 'y'})
#show the data
df
```

```python
#Get the last 20 rows of data and store them into a new variable
last = df[len(df)-20:]
last
```

```python
#Get all of the rows in the data set except for the last 20 rows of data
df = df[:-20]
df
```

```python
#getting the Facebook prophet library
from fbprophet import Prophet
#creating the Prophet object (model)
fbp = Prophet(daily_seasonality = True)
#fit/train the model
fbp.fit(df)
future = fbp.make_future_dataframe(periods=365)
forecast = fbp.predict(future)
```

```python
#import the facebooks prophet plotting libraies
from fbprophet.plot import plot_plotly
#plot the data
plot_plotly(fbp,forecast)
```

```python
#show the models prediction for 10-15-2021
forecast[forecast.ds == '2022-10-15']['yhat']
```
